## Correlation

Let's begin our investigation into the possible impacts of smoke on Richland by finding the correlation between the airquality due to smoke and the hospitalization rate. Because the healtchare data only has data points from 2012 - 2020 (excluding 2015), we will limit our examination to these years. Recall that my AQI data come from very near Richland. We will assume that the patient discharge data and AQI is representative of Richland, WA.

First, let's plot the change in population of Benton County. Before we do that, however, we need to grab 2012 - 2020 (again excluding 2015) from the population estimate data to align with the healthcare data.

```python
# filter to the relevant years
df_Benton_pop_est = df_Benton_pop_est[df_Benton_pop_est['Year'] > 2011]
df_Benton_pop_est = df_Benton_pop_est[df_Benton_pop_est['Year'] < 2021]
df_Benton_pop_est = df_Benton_pop_est.reset_index(drop=True)

df_Benton_pop_est = df_Benton_pop_est.drop(index=3)
df_Benton_pop_est = df_Benton_pop_est.reset_index(drop=True)
df_Benton_pop_est
```

```python
# plot the data
x = df_Benton_pop_est['Year']
y = df_Benton_pop_est['Population Estimate']
plt.plot(x, y)

plt.xlabel('Year')
plt.ylabel('Population Estimate')
plt.title('Benton Population')

plt.show()
```

Let's look at the change in patient discharges

```python
# -------------------------- referenced In [9] from part 1 ------------------------#
# plot a time series of annual discharges

# healthcare
x1 = df_Benton_annual_discharges['Year']
y1 = df_Benton_annual_discharges['Total_Discharges']
plt.plot(x1, y1, label = 'patient discharges')

# add a legend
plt.legend()

plt.xlabel('Year')
plt.ylabel('Patient Discharges')
plt.title('Total Annual Patient Discharges')

# plot
plt.show()
```

Also, because population fluctuates, we need to control for that when looking at how the annual patient discharges for selected respiratory issues changes over time. Let's divide the number of discharges each year by the population estimate of Benton County that year to control for population variation and then plot the ratio.

```python
# calculate ratio
df_discharge_ratio = pd.DataFrame()
df_discharge_ratio['Year'] = df_Benton_pop_est['Year']
df_discharge_ratio['Discharge Ratio'] = df_Benton_annual_discharges['Total_Discharges'] / df_Benton_pop_est['Population Estimate']
df_discharge_ratio
```

Let's convert the discharge ratio from discharges per person in Benton to discharges per 1,000 people

```python
df_discharge_ratio['Discharge Ratio'] = df_discharge_ratio['Discharge Ratio'] * 1000
df_discharge_ratio
```

Again, let's plot the results

```python
# -------------------------- referenced In [9] from part 1 ------------------------#
# plot a time series of annual discharges per 1000 people
x = df_discharge_ratio['Year']
y = df_discharge_ratio['Discharge Ratio']
plt.plot(x, y)

plt.xlabel('Year')
plt.ylabel('Patient Discharges')
plt.title('Annual Patient Discharges per 1000 People')

# plot
plt.show()
```

Looking at the plot above, the trend appears to be simillar to the patient discharge plot earlier where we did not control for population. It is interesting the number of discharges drops in 2020. I would have expected the number to increase significantly due to Covid-19 being a respiratory disease. Still, let's remove it from consideration as it still may be an outlier due to Covid-19.

```python
df_discharge_ratio = df_discharge_ratio[df_discharge_ratio['Year'] != 2020]
df_discharge_ratio = df_discharge_ratio.reset_index(drop=True)
df_discharge_ratio
```

Let's plot the figure above again

```python
# -------------------------- referenced In [9] from part 1 ------------------------#
# plot a time series of annual discharges per 1000 people
x = df_discharge_ratio['Year']
y = df_discharge_ratio['Discharge Ratio']
plt.plot(x, y)

plt.xlabel('Year')
plt.ylabel('Patient Discharges')
plt.title('Annual Patient Discharges per 1000 People')

# plot
plt.show()
```

Now, let's observe the AQI trends. First, let's get the aqi data for 2012 - 2020 excluding 2015 and remove 2020 from consideration here as well, to align with the discharge data.

```python
df_annual_relevant_aqi = df_annual_aqi[df_annual_aqi['Year'] > 2011]
df_annual_relevant_aqi = df_annual_relevant_aqi[df_annual_relevant_aqi['Year'] < 2021]
df_annual_relevant_aqi = df_annual_relevant_aqi.reset_index(drop=True)

# remove the entry for 2015
df_annual_relevant_aqi = df_annual_relevant_aqi.drop(index=3).reset_index(drop=True)

# drop useless column
df_annual_relevant_aqi = df_annual_relevant_aqi.drop(columns=['Unnamed: 0'], axis=1)

# drop 2020
df_annual_relevant_aqi = df_annual_relevant_aqi[df_annual_relevant_aqi['Year'] != 2020]
df_annual_relevant_aqi = df_annual_relevant_aqi.reset_index(drop=True)
df_annual_relevant_aqi

df_annual_relevant_aqi
```

Now, let's plot the AQI.

```python
# -------------------------- referenced In [9] from part 1 ------------------------#
# plot a time series of annual discharges

# aqi
x = df_annual_relevant_aqi['Year']
y = df_annual_relevant_aqi['AQI']
plt.plot(x, y,)

plt.xlabel('Year')
plt.ylabel('AQI')
plt.title('Annual Average AQI')

# plot
plt.show()
```

Now, let's look at the correlation between the AQI data and annual total patient discharges for Benton County. We begin by constructing a scatter plot. I referenced [2]. to learn that a scatter plot is a useful way to visualize correlation - as well as supplemented my knowledge of how to create one.

```python
# -------------------------- referenced In [8] from part 1 ------------------------#
# create a scatter plot
# I referenced https://www.w3schools.com/python/matplotlib_scatter.asp for how to create
# a scatte plot
x = df_annual_relevant_aqi['AQI']
y = df_discharge_ratio['Discharge Ratio']
plt.scatter(x, y)

plt.xlabel('AQI')
plt.ylabel('Patient Discharges')
plt.title('Relationship Between AQI and Patients Discharged per 1K People')
plt.show()
```

The data is relatively sparse and there is no obvious correlation. Let's caculate the correlation numerically to confirm this.

```python
# -------------------------- referenced In [10] from part 1 ------------------------#
# calculate the correlation
# I learned how to do this from: https://pandas.pydata.org/docs/reference/api/pandas.Series.corr.html
x.corr(y)
```

This result basically confimrs what we interpreted from the graph. There is a very slight negative correlation - essentially no correlation between the AQI and number of patients discharged per 1000 people. These findings surprise me, as I expected to see a positive correlation between AQI and the number of patient discharges.

## Modeling

Because we determined that there is virtually no corellation between AQI and patient discharges, above, I do not think it makes sense to attempt to predict the future number of discharges based on AQI. 

There is a second reason for not doing predictive modeling in this context. I neglected to realize when I wrote my part 2 extension plan that performing predictive modeling would be quite challenging due to having limited data on which to predict upon, as that information is in the future and we don't have it yet. For instance, say I wanted to train a simple linear regression model where the independant variables are the year and annual AQI of each year, and the response variable was number of hospitalizations per 1,000 people. Well, once I train this model, I would want to get predictions on what future years would look like. However, I do not have future values for AQI. Thus, I cannot predict future levels of hospital discharges very easily.

I conducted a brief survey of several relevant articles ([7], [8], and [9]) to see if they did any predictive modeling. All three articles utilized historical data (or esimates) to determine if the smoke had a statistically measurable impact on peoples' health. That is, they are not performing predictive modeling (at least not that I noticed), but statistically analyses on the association of variables and/or levels of risk. This confirms my suspicion that I probably should not have aimed for predictive modeling in my extension plan. 

Although there are strong reasons against predictive modeling in this context, for the sake of satisfying my initial plan, I will produce a predictive model. However, I will state upfront that I have little confidence in this approach, and it would need further examination for mathematical validity before it is used.

Now, let's predict the hospitalization rate for Richland for the following 25 years (i.e. from 2020 - 2045). To generate the future AQI that we do not have, we will first predict the future air quality, and then use those results to get predictions for future hospital discharge rates. Let's start with predicting future AQI.

```python
import statsmodels.api as sm

# ---------------- based off of cell In [6] from [1] and on [3] ----------------------- #

# get training data
X_train = np.array(df_annual_relevant_aqi['Year'])
Y_train = np.array(df_annual_relevant_aqi['AQI'])

# TBD not sure why next line works but needed when x is single feature
X_train = np.reshape(X_train, (-1, 1))

# add the constant term
X_train = sm.add_constant(X_train)

# train the model
result = sm.OLS(Y_train, X_train).fit()

# print the summary of results
print(result.summary())
```

According to source [5], generally higher R-squared values (closer to 1) are better. As we see from the results, the R-squared value is extremely low, suggesting that this model is a terrible fit for the data. Let's visualize the relationship between year and AQI.

```python
x = df_annual_relevant_aqi['Year']
y = df_annual_relevant_aqi['AQI']

plt.scatter(x,y)
plt.xlabel('Year')
plt.ylabel('AQI')
plt.title('Year vs AQI')
plt.show()
```

Now, let's predict AQI with the model, even though it is a poor fit - for the sake of the exercise.

```python
# ---------------- based off of cell In [6] from [1] and on [6] ----------------------- #

# get predictions
years = list(range(2020, 2045 + 1))
X_preds = years

# --------------------- TBD I consulted sources [7]. and [8]. and realized I still needed to add a
# constant term
X_preds = sm.add_constant(X_preds)

Y_preds = result.predict(X_preds)

# construct dataframe
df_future_aqi_pred = pd.DataFrame()
df_future_aqi_pred['Year'] = years
df_future_aqi_pred['AQI_Prediction'] = list(Y_preds)

df_future_aqi_pred
```

Let's plot the predicted AQI

```python
x = df_future_aqi_pred['Year']
y = df_future_aqi_pred['AQI_Prediction']
plt.plot(x, y)

plt.title('AQI Predictions')
plt.xlabel('Year')
plt.ylabel('Predicted AQI')
plt.show()
```

We observe that the AQI is predicted to slowly increase over the next 25 years. Now, we train a model to predict future hospital discharges. 

```python
# ---------------- TBD based off of cell In [6] from [1] and on [3] ----------------------- #

# get training data
X_train = np.array(df_annual_relevant_aqi)
Y_train = np.array(df_discharge_ratio['Discharge Ratio'])

# add the constant term
X_train = sm.add_constant(X_train)

# train the model
result = sm.OLS(Y_train, X_train).fit()

# print the summary of results
print(result.summary())
```

After training this new model, we see that the R-squared value is actually decent - certainly better than I expected it to be after finding the correlation and scatter plot earlier. I am not sure why I am observing this here.

Now, let's go ahead and make hospital discharge predictions for 2020 through 2045 based upon the future AQI values I predicted earlier. Again, I am not sure of the statistical implications of running a prediction on another prediction. 

```python
# ---------------- based off of cell In [6] from [1] and on [6] ----------------------- #

# get predictions
X_preds = np.array(df_future_aqi_pred)

# --------------------- TBD I consulted sources [7]. and [8]. and realized I still needed to add a
# constant term
X_preds = sm.add_constant(X_preds)

Y_preds = result.predict(X_preds)

# construct dataframe
df_future_discharges_preds = df_future_aqi_pred
df_future_discharges_preds['Patient_Discharge_Prediction'] = list(Y_preds)
df_future_discharges_preds
```

Finally, let's plot this result

```python
x = df_future_discharges_preds['Year']
y = df_future_discharges_preds['Patient_Discharge_Prediction']
plt.plot(x, y)

plt.title('Predicted Patient Discharges (Per 1K People)')
plt.xlabel('Year')
plt.ylabel('Discharges')
plt.show()
```
