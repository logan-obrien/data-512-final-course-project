## Locate Fires Closest to City:

Next, according to the assignment instructions, my smoke estimate should only consider fires within 1250 miles of my assigned city, Richland, WA. So, let us filter for fires within this range.

The first step is to write several functions for calculating the distance between the fires and my city. The first function is defined in the cell below and taken directly from [3].

```python
#
#    Transform feature geometry data
#
#    The function takes one parameter, a list of ESRI:102008 coordinates that will be transformed to EPSG:4326
#    The function returns a list of coordinates in EPSG:4326
def convert_ring_to_epsg4326(ring_data=None):
    converted_ring = list()
    #
    # We use a pyproj transformer that converts from ESRI:102008 to EPSG:4326 to transform the list of coordinates
    to_epsg4326 = Transformer.from_crs("ESRI:102008","EPSG:4326")
    # We'll run through the list transforming each ESRI:102008 x,y coordinate into a decimal degree lat,lon
    for coord in ring_data:
        lat,lon = to_epsg4326.transform(coord[0],coord[1])
        new_coord = lat,lon
        converted_ring.append(new_coord)
    return converted_ring
```

After defining the helper function above, we now write the function that does the heavy lifting and actually computes the distances. Again, the code and comments are copied directly from [3]. 

```python
#    
#    The function takes two parameters
#        A place - which is coordinate point (list or tuple with two items, (lat,lon) in decimal degrees EPSG:4326
#        Ring_data - a list of decimal degree coordinates for the fire boundary
#
#    The function returns the average miles from boundary to the place
#
def average_distance_from_place_to_fire_perimeter(place=None,ring_data=None):
    # convert the ring data to the right coordinate system
    ring = convert_ring_to_epsg4326(ring_data)    
    # create a epsg4326 compliant object - which is what the WGS84 ellipsoid is
    geodcalc = Geod(ellps='WGS84')
    # create a list to store our results
    distances_in_meters = list()
    # run through each point in the converted ring data
    for point in ring:
        # calculate the distance
        d = geodcalc.inv(place[1],place[0],point[1],point[0])
        distances_in_meters.append(d[2])
    #print("Got the following list:",distances_in_meters)
    # convert meters to miles
    distances_in_miles = [meters*0.00062137 for meters in distances_in_meters]
    # the esri polygon shape (the ring) requires that the first and last coordinates be identical to 'close the region
    # we remove one of them so that we don't bias our average by having two of the same point
    distances_in_miles_no_dup = distances_in_miles[1:]
    # now, average miles
    average = sum(distances_in_miles_no_dup)/len(distances_in_miles_no_dup)
    return average
```

After creating these functions, the next step is to find the longitude and latitude for Richland. I used Google to find this information [5] and modified the structure of the CITY constant that was written in [3] to fit my needs.

```python
# Define the data for my city, Richland WA:
CITY = {
    'city' : 'Richland',
    'latlon' : [46.28569070, -119.28446210]
}
```

At this point, we can iterate through all the loaded features (fires) and filter out those that do not meet the assignment specifications. Specifically, the fires must occurr between 1963 and 2023 and be within 1250 miles of Richland Washington. To do this, I reuse and modify a loop that Dr. McDonald wrote in [3], so the following code cell has a blend of my code and comments and his.

One issue I encountered is that for a subset of the features, the have 'curveRings' geometries instead of 'rings'. The professor said on Slack that he found only 35 such features and that he was fine if we ignore them. To accomplish this, I created some try-catch clauses and assign a value of '-1' to my distance variable if the feature does not have a 'rings' geometry. Also, when I perform the filtering, I filter out these features so that they are ignored, by checking for the '-1' value for distance.

Also, during my work on the next cell, I referenced several online sources [6], [7], and [8], as well as reusing and modifying a code snippet I wrote in another assignment for this class [9].

```python
fires_attributes_in_range_list = []
num_curveRings = 0

for wf_feature in tqdm(feature_list):
    wf_id = wf_feature['attributes']['OBJECTID']
    
    
    wf_year = wf_feature['attributes']['Fire_Year']
    #
    #     Compute using the average distance to all points on the perimeter
    #
    # Try to extract the first ring
    try:
        ring_data = wf_feature['geometry']['rings'][0]
    except: #if no ring data, check curveRings
        try:
            ring_data = wf_feature['geometry']['curveRings'][0]
        except:
            print("Error getting ring data for OBJECTID {}".format(wf_id))
            distance = -1
        else: #if it is a curveRing, increment counter
            distance = -1
            num_curveRings = num_curveRings +1
    else: #if no error
        distance = average_distance_from_place_to_fire_perimeter(CITY['latlon'],ring_data)
    
    # Make sure the fire occurred within the required time range and distance from Richland
    # Also, ensure that it had a valid distance
    if (1963 <= wf_year) and (-1 < distance <= 1250):        
        #-----------TBD add more fields as needed
        fire_attributes = wf_feature['attributes']
        
        # Add distance to Richland
        fire_attributes['distance_to_richland'] = distance
        
        fires_attributes_in_range_list.append(fire_attributes)

#-------- code snippet from [9] -----------
# Save to JSON file - Reference: https://www.geeksforgeeks.org/writing-to-file-in-python/
my_file = open(r"./data/relevant_fires.json", "w")
my_file.write(json.dumps(fires_attributes_in_range_list, indent=4)) # Source: borrowed part of a line of code from Prof example
my_file.close()
#------------------------------------------

print("Done!")
print("There were {} fires skipped due to having curveRings".format(num_curveRings))
```
